#### Count Of Frequency Encoding

```python
train_set= pd.read_csv(r'C:\Users\hp\adult.csv')
```

```python
train_set.head()
```

```python
columns=['workclass','education','marital-status','occupation','relationship','race','sex','native-country']
```

```python
df = train_set.iloc[: , [1,3,5,6,7,8,9,13]].copy() 
```

```python
df.head()
```

```python
for feature in df.columns[:]:
    print(feature,":",len(df[feature].unique()),'labels')
```

```python
df['native-country'].value_counts()
```

```python
country_map= df['native-country'].value_counts().to_dict()
```

```python
df['native-country']=df['native-country'].map(country_map)
```

```python
df.head(30)
```

### Advantages
1.Easy to use
2.Not increasing any feature space

### Disadvantages
1.It willl provide the same weight if the frequencies are same (eg. two classes have same number)


### Target Guided Ordinal Encoding

##### 1. Ordering the labels according to the target
##### 2. Replace the labels by the joint probability of being 1 or 0

```python
df=pd.read_csv(r'C:\Users\hp\Downloads\Titanic\titanic.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
df['Cabin'].fillna("Missing",inplace=True)
```

```python
df.head()
```

```python
df['Cabin'].value_counts().to_dict()
```

```python
df['Cabin']=df['Cabin'].astype(str).str[0]
```

```python
df.head()
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels= df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
enumerate(ordinal_labels,0)
```

```python
ordinal_labels2= {k:i for i,k in enumerate(ordinal_labels,0)}
ordinal_labels2
```

```python
df['Cabin_ordinal_labels']= df['Cabin'].map(ordinal_labels2)
df.head()
```

### Mean Encoding

```python
mean_ordinal=df.groupby(['Cabin'])['Survived'].mean()
```

```python
df['mean_ordinal_encod']=df['Cabin'].map(mean_ordenal)
df.head()
```
